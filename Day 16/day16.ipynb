```python
fh = open("input.txt")
contents = fh.read()
contents
```

```python
hex_to_bin = {}
hex_to_bin['0'] = '0000'
hex_to_bin['1'] = '0001'
hex_to_bin['2'] = '0010'
hex_to_bin['3'] = '0011'
hex_to_bin['4'] = '0100'
hex_to_bin['5'] = '0101'
hex_to_bin['6'] = '0110'
hex_to_bin['7'] = '0111'
hex_to_bin['8'] = '1000'
hex_to_bin['9'] = '1001'
hex_to_bin['A'] = '1010'
hex_to_bin['B'] = '1011'
hex_to_bin['C'] = '1100'
hex_to_bin['D'] = '1101'
hex_to_bin['E'] = '1110'
hex_to_bin['F'] = '1111'
```

```python
bits = []
for c in contents:
    bits.append(hex_to_bin[c])
```

```python
"".join(bits)
```
